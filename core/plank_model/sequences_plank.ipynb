```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("plank_model/data/plank_dataset.csv")
df.head()

# Colonnes des landmarks
lm_cols = [c for c in df.columns if c != "label"]
print("Nombre de colonnes landmarks :", len(lm_cols))
```

```python
SEQ_LEN = 50   # nombre de frames par séquence
STEP = 10      # pas entre les séquences

X, y = [], []

for i in range(0, len(df) - SEQ_LEN, STEP):
    seq = df[lm_cols].iloc[i:i+SEQ_LEN].values.astype(np.float32)
    X.append(seq)
    y.append(0)  # classe unique Plank pour l'instant

X = np.array(X, dtype=np.float32)
y = np.array(y, dtype=np.int64)

print("Shape X :", X.shape)
print("Shape y :", y.shape)
```
